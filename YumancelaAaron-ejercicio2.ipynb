**Objetivo de la práctica**
* Comprender el modelo vectorial como base para representar documentos y consultas.
* Calcular la matriz TF-IDF para el corpus data/01_corpus_turismo_500.txt

```python
 # Cargamos el dataset

import pandas as pd
import time

# Se define la ruta del archivo
path = '/kaggle/input/corpus/01_corpus_turismo_500.txt'

# Hacemos que se lean todas las líneas 
with open(path, 'r', encoding='utf-8') as f:
    documentos = f.readlines()

print(f" Corpus cargado correctamente. Total de documentos: {len(documentos)}")
```

```python
# Ejercicio 2 - Modelo Vectorial y TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# 1. Creamos el vectorizador TF-IDF

vectorizador = TfidfVectorizer(
    lowercase=True,         # convertimos a minúsculas
    stop_words=None,        # no eliminamos palabras por ahora
    norm='l2'               # normaliza los vectores (magnitud 1)
)

# 2. Calculamos la matriz TF-IDF del corpus

tfidf_matriz = vectorizador.fit_transform(documentos)

# 3. Convertimos la matriz a DataFrame (opcional, para verla mejor)

df_tfidf = pd.DataFrame(
    tfidf_matriz.toarray(), 
    columns=vectorizador.get_feature_names_out()
)

print("Matriz TF-IDF calculada")
print(f"Dimensiones: {df_tfidf.shape[0]} documentos × {df_tfidf.shape[1]} términos\n")

# Mostrar una vista parcial de la matriz
#df_tfidf.head()


pd.set_option('display.max_columns', None)  # Mostramos todas las columnas
pd.set_option('display.max_rows', 10)      # Mostramos solo las primeras 10 filas
print(df_tfidf.head(10))                    # Muestramos 10 documentos completos
```
